```python
def generate_title(subjects):
    prompt= f"give me a title for a book that has something to do with {subjects[0]} and {subjects[1]}"
    
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "system", "content": (prompt)}]
    )
    
    return response.choices[0].message.content
```

```python
subjects = ['IT', 'Coding', 'Programming', 'unicorns', 'frontend', 'backend', 'javascript', 'Stakeholder Niek']
```

```python
rating_data = []

for user in users:
    books_rated = random.sample(books, 20)
    
    for book in books_rated:
        rating = random.choice([1,2,3,4,5])
        
        rating_data.append({
            'User': user,
            'Book': book,
            'Rating': rating
        })
```

```python
def generate_review(book, rating):
    prompt = f"write a brief review for an IT book with a rating of {rating}. Do not explicitly mention the rating or title. The book is called: "
    
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "system", "content": (prompt + book)}]
    )
    
    return response.choices[0].message.content
```

```python
i = 0
for review in rating_data:
    i += 1
    print(i)
    review['Review'] = generate_review(review['Book'], review['Rating'])
```
